```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_images = os.listdir('../data/train/')
test_images = os.listdir('../data/val/')
```

```python
print(f'No of training images {len(train_images)}')
print(f'No of validation images {len(test_images)}')
```

```python
for i in range(5):
    index = np.random.randint(low=0, high= len(train_images)-1)
    image = cv2.cvtColor(cv2.imread(os.path.join('../data/train/', train_images[index])), cv2.COLOR_BGR2RGB)
    image, label = image[:, :256, :], image[:, 256:, :] 
    fig, axes = plt.subplots(1, 2, figsize=(10, 6))
    axes[0].imshow(image)
    axes[0].set_xticks([])
    axes[0].set_yticks([])
    axes[1].imshow(label)
    axes[1].set_xticks([])
    axes[1].set_yticks([])
    plt.tight_layout()
plt.show()
```

```python
for i in range(5):
    index = np.random.randint(low=0, high= len(test_images)-1)
    image = cv2.cvtColor(cv2.imread(os.path.join('../data/train/', test_images[index])), cv2.COLOR_BGR2RGB)
    image, label = image[:, :256, :], image[:, 256:, :] 
    fig, axes = plt.subplots(1, 2, figsize=(10, 6))
    axes[0].imshow(image)
    axes[0].set_xticks([])
    axes[0].set_yticks([])
    axes[1].imshow(label)
    axes[1].set_xticks([])
    axes[1].set_yticks([])
    plt.tight_layout()
plt.show()
```
